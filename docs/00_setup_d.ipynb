# Setup


## 1. Installation der Anaconda-Distribution

https://www.continuum.io/downloads - Installer für Python 3.6 herunterladen und Software installieren.


## 2. Virtual Environment einrichten

Je nach Betriebssystem in Konsole, Terminal oder Shell:

### Virtual Environment anlegen:
```
> conda create --name legalrobots
```

Oder kurz:
```
> conda create -n legalrobots
```

### Virtual Environment aktivieren:

* **Windows:**
```
> activate legalrobots
```

* **Linux/Mac:**
```
> source activate legalrobots
```


> **Hinweis:** 

>Die vorhandenen Virtual Environments lassen sich mit `conda env list` oder `conda info --envs` anzeigen. 
Der Befehl `conda deactivate legalrobots` (Windows) bzw. `source deactivate legalrobots` (Linux/Mac) führt in das Root Environment zurück. 
Nähere Informationen im [conda cheat sheet](https://conda.io/docs/using/cheatsheet.html).


### Conda-Integration für Notebooks installieren:

```
> conda install -c anaconda nb_conda
```

## 3. Packages installieren

Benötigte Packages werden im aktivierten Virtual Environment `legalrobots` installiert.
Hier gibt es grundsätzlich zwei Möglichkeiten:

* Packages, die auch über `conda` verfügbar sind ([Liste der verfügbaren Packages](https://docs.continuum.io/anaconda/pkg-docs)):
```
> conda install package_name
```

* Packages, die über `pip` verfügbar sind:
```
> pip install package_name
```

Man kann mehrere Packages gleichzeitig installieren, indem man deren Package-Namen im Installationsbefehl getrennt durch ein Leerzeichen hintereinanderschreibt.
Falls wir eine bestimmte Version eines Packages benötigen, können wir diese (mit `=` und einigen verwandten Zeichen) an den Package-Namen anhängen, beispielsweise `package_name=2.3` für Version(en) 2.3 von `package_name`.

Die Packages, die wir im Rahmen von LegalRobots brauchen, installieren wir kollektiv via `pip`, da nicht alle Packages via `conda` verfügbar sind:

```
> pip install ipython-sql prettytable pandas cufflinks plotly
```
